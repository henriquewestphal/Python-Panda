```python
import pandas as pd
```

```python
dados = pd.read_csv('dados/aluguel_resedencial.csv', sep=';')
```

```python
dados.head()
```

```python
dados['Valor'].mean()
```

```python
bairros = dados['Bairro'].drop_duplicates()
selecao = dados['Bairro'].isin(bairros)
dados = dados[selecao]
```

```python
dados
```

```python
group_bairro = dados.groupby('Bairro')
```

```python
type (group_bairro)
```

```python
group_bairro.groups
```

```python
for bairro, data in group_bairro:
    print('{} -> {}'.format(bairro,data.Valor.mean()))
```

```python
group_bairro['Valor','Condominio'].mean().round(2)
```

## Estatísticas Descritivas

```python
 group_bairro['Valor'].describe().round(2)
```

```python
group_bairro['Valor'].aggregate(['min','max']).rename(columns = {'min': 'Mínimo', 'max':'Máximo'})
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.rc('figure', figsize = (20,15))
```

```python
fig = group_bairro['Valor'].mean().plot.bar(color = 'blue')
fig.set_ylabel('Valor do Aluguel')
fig.set_title('Valor Médio do Aluguel por Bairro', {'fontsize':22})
```
